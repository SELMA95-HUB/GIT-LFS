## Quelle est la proportion de chacun des genres botaniques par rapport au total ?

```python
# calcul de total arbre
tot_arbre=[]
for x in data:
    tot_arbre.append(x["CATEGORIE_DESC"])
    y=len(tot_arbre)
print(y)
```

```python
# pourcentage
proportion_genre={}
for genre, nombre in compte_genre.items():
    proportion = round((nombre/y)*100,2)
    proportion_genre[genre]=proportion
```

```python
pourc_genre_decr = sorted(proportion_genre.items(), key = lambda item:item[1], reverse=True)
print(pourc_genre_decr)
```
